```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("LivingPlanetIndex_2022_PublicData/LPD2022_public.csv", dtype='str', index_col=0, header=0)
```

```python
data["Region"].value_counts()
```

```python
LAC = data[data["Region"]=="Latin America and Caribbean"]
```

```python
def count_per_year(df, col_name, field, start_year=1950, end_year=2020):
    return df[df[col_name]==field].loc[:, str(start_year):str(end_year)].notna().sum()

def ordered_counts(df, col_name, start_year=1950, end_year=2020):
    counts = pd.DataFrame()
    for field in df[col_name].unique():
        count = count_per_year(df, col_name, field, start_year, end_year)
        counts = counts.assign(**{f"{field}: {count.sum()}": count})
    sorted_columns = sorted(counts.columns, key=lambda x: counts.loc[:, x].sum(), reverse=True)
    return counts[sorted_columns]
```

```python
plt.figure(figsize=(20,6))
count_per_year(data, "Region", "Latin America and Caribbean").plot()
plt.title('Total Number of Data Entries per Year in Latin America and Caribbean', fontsize=20)
plt.xlabel('Year', fontsize=16)
plt.ylabel('# Data Entries', fontsize=16)
plt.xticks()
plt.grid()
plt.show()
```

```python
def plot_data_counts(counts, name, top=10):
    counts.iloc[:, :top].plot(figsize=(20,6))
    plt.grid()
    plt.legend(loc='upper left', fontsize=14, title=f"{name}: Total Count", title_fontsize=16)
    plt.title(f"Amount of Data for each '{name}': Showing {top} with Most Total Data", fontsize=20)
    plt.xlabel("Year", fontsize=16)
    plt.ylabel("# Data Entries", fontsize=16)
    plt.show()
```

```python
unit_counts = ordered_counts(LAC, "Units")
plot_data_counts(unit_counts, "Unit")
```

The most common unit of measurement in the Latin American and Caribbean dataset is by far "average number of individuals". However, almost all of this data appears to be in the 2000s. The data could still be useful, but it is important to check what data is available is the timeframe before and during the 1990s, which is when the Convention of Biological Diversity began to take effect.

```python
unit_counts_pre2000s = ordered_counts(LAC, "Units", start_year=1970, end_year=2000)
plot_data_counts(unit_counts_pre2000s, "Unit")
```

The next step will be to determine what species and regions have the most data, and then the overlap among species, units, and regions. In most cases, each individual study in the data set will be for a set of species in a certain area and using a certain type of unit. After discovering which studies contain useful data, the data from those studies can be extracted into a new data set to use for this research project.

```python
genus_counts = ordered_counts(LAC, "Genus")
plot_data_counts(genus_counts, "Genus")
```
